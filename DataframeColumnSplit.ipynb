```python
def num_cat_col(df):
    '''
    Splits the columns of the dataframe into categorical and numerical columns and returns them as 2 separate lists.
    Columns with IDs in them are added to the list of categorical columns.
    '''
    num_col = []
    cat_col = []
    for column in df.columns:
        if df[column].dtypes == 'object' or '_id' in column:
            cat_col.append(column)
        else:
            num_col.append(column)
    return cat_col, num_col

cat_col, num_col = num_cat_col(patient_df)
```
